Descargar e instalar la version de tesseract que puedan usar:

https://github.com/UB-Mannheim/tesseract/wiki

- tesseract-ocr-w32-setup-v5.0.0-alpha.20190708.exe (32 bit) and
- tesseract-ocr-w64-setup-v5.0.0-alpha.20190708.exe (64 bit) resp.

Luego de instalar sigue el <b>pip install pytesseract</b>

Establecer la ruta de tesseract en el script antes de llamar a <b>image_to_string</b>:

pytesseract.pytesseract.tesseract_cmd = r<b>'RUTA DONDE SE INSTALO TESSERACT/tesseract''</b>

En mi caso, se instalo acá:

<img src="1.png">

```python
import pytesseract
import cv2
from PIL import Image
from PIL import Image
```

```python
pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract-OCR/tesseract'
```

```python
im = Image.open("texto/0.jpg")
texto = pytesseract.image_to_string(im)
print(texto)
Image.open("texto/0.jpg")
```
