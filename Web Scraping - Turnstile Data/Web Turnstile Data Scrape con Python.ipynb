```python
import requests
import urllib.request
import pandas as pd
import io
from time import time
from datetime import datetime, timedelta
from bs4 import BeautifulSoup

timeFinifih = 0
start_time = time()

url = "http://web.mta.info/developers/turnstile.html"

response = requests.get(url)

instancia = 1

print('\n' + str(instancia) + ') Verificando conectividad a la url solicitada:')
instancia = instancia + 1

if response.status_code == 200:    
    print('•  Acceso Ok!')
    soup = BeautifulSoup(response.text, "html.parser")
    enlaces = soup.findAll('a')
    url_base = 'http://web.mta.info/developers/'
    txt_info = []
    print('\n' + str(instancia) + ') Verificando sub urls')
    for x in range(0, len(enlaces)):
        data = soup.findAll('a')[x]
        direccion = str(data).find('href')
        if direccion != -1:
            txt_file = data['href'].find('.txt')
            if txt_file != -1:
                folder = data['href'].find('data/nyct/turnstile/')
                if folder != -1:
                    txt_info.append(url_base + data['href'])
    print('•  Accesos Ok!')
    instancia = instancia + 1
    print('\n' + str(instancia) + ') El array txt_info tiene un total de ' + str(len(txt_info)) + ' enlaces')
    instancia = instancia + 1
    print('\n' + str(instancia) + ') Generando df base para obtener nombre de columnas ...')
    data = requests.get(txt_info[0]).content
    df = pd.read_csv(io.StringIO(data.decode('utf-8')))
    columnas = df.columns.tolist()
    instancia = instancia + 1
    print('\n' + str(instancia) + ') Columnas obtenidas: ' + str(columnas))
    df_final = pd.DataFrame(columns = columnas)
    enlaces_scrapeados = 10
    instancia = instancia + 1
    print('\n' + str(instancia) + ') Creando df_final con información de ' + str(enlaces_scrapeados) + ' enlaces seleccionados ...')
    for x in range (0, 10):
        nueva_data = requests.get(txt_info[x]).content
        nuevo_df = pd.read_csv(io.StringIO(nueva_data.decode('utf-8')))
        df_final = pd.concat([df_final, nuevo_df], axis=0)
        print('     Df generado enlace ' + str(x + 1) + ' de ' + str(enlaces_scrapeados) + ': Shape ' + str(df_final.shape))
    instancia = instancia + 1
    print('\n' + str(instancia) + ') df_final creado!')
    instancia = instancia + 1
    print('\n' + str(instancia) + ') df_final shape:' + str(df_final.shape))
    timeFinifih += (time() - start_time)
    instancia = instancia + 1
    print('\n' + str(instancia) + ') Tiempo total de gestión: ' + str(timedelta(seconds=timeFinifih)))
else:
    print('No se pudo conectar a la URL')
```
