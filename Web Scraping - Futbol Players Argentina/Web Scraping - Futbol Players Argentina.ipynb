```python
import requests
import urllib.request
import pandas as pd
import re
import json
from time import time
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

palette = ["#3a9679", "#a7d129","#ff3d00", "#00e676", "#ad1457", "#f09c67", "#257aa6", 
           "#ffab00", "#e16262", "#263238"]

sns.palplot(sns.color_palette(palette))
```

```python
path_file_dest = r'futbol.json'
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (Platform; Security; OS-or-CPU; Localization; rv:1.4) Gecko/20030624 Netscape/7.1 (ax)'})
```

```python
def write_doc_json(content, mode):
    own_file = open(path_file_dest, mode, encoding='utf-8')
    own_file.write(content.decode('utf-8'))
    own_file.close()
    print('\n•  El archivo json fue sobre-escrito con exito!')
```

```python
if response.status_code == 200:
    
    info = []
    soup = BeautifulSoup(response.text, "html.parser")
    data = soup.findAll('div', attrs={'class': 'boxhome boxhome-2col'})
    
    for x in range (0, len(data)):
        
        equipo = data[x].find('span', {'class': 'titlebox'}).get_text().lstrip()
        tabla = data[x].find('tbody')
        jugadores = tabla.findAll('tr')
        
        print('________________________________________________________________')
        print('')
        print(equipo)
        print('')
        
        for i in range (0, len(jugadores)):
            datos_jugador = []
            nombre = jugadores[i].findAll('td')[0].find('strong').get_text()
            direccion_jugador = 'https://www.resultados-futbol.com' + jugadores[i].find('td').find('a')['href']
            
            response_dos = requests.get(direccion_jugador, headers={'User-Agent': 'Mozilla/5.0 (Platform; Security; OS-or-CPU; Localization; rv:1.4) Gecko/20030624 Netscape/7.1 (ax)'})
           
            print(nombre)
            
            if response_dos.status_code == 200:
                info_dos = []
                soup_dos = BeautifulSoup(response_dos.text, "html.parser")
                data_dos = soup_dos.findAll('dd')
                
                for l in range (0, 9):
                
                    datos_jugador.append(data_dos[l].get_text().rstrip().lstrip())

                individuo = {
                    'Equipo' : equipo,
                    'Nombre_Completo' : datos_jugador[1],
                    'Edad' : datos_jugador[2],
                    'Fecha_Nacimiento' : datos_jugador[3],
                    'Provincia' : datos_jugador[4],
                    'Pais' : datos_jugador[5],
                    'Nacionalidad' : datos_jugador[6],
                    'Posicion' : datos_jugador[7],
                    'Altura' : datos_jugador[8]
                    }
                
                info.append(individuo)
        print('')
            
    content = json.dumps(info, indent=4, sort_keys=True, ensure_ascii=False).encode('utf-8')
    write_doc_json(content ,'w+')
    data = pd.read_json(path_file_dest, encoding = 'UTF-8')
    print('•  Se leyo el json y se creo el dataframe data')
    
else:
    print('No se pudo conectar')
```

```python
dataFinal = pd.read_json('futbol.json', encoding='utf8')

dataFinal.Equipo = dataFinal.Equipo.apply(lambda x: bytes(x,'iso-8859-1').decode('utf-8'))
dataFinal.Nombre_Completo = dataFinal.Nombre_Completo.apply(lambda x: bytes(x,'iso-8859-1').decode('utf-8'))
dataFinal.Provincia = dataFinal.Provincia.apply(lambda x: bytes(x,'iso-8859-1').decode('utf-8'))
dataFinal.Pais = dataFinal.Pais.apply(lambda x: bytes(x,'iso-8859-1').decode('utf-8'))
dataFinal.Nacionalidad = dataFinal.Nacionalidad.apply(lambda x: bytes(x,'iso-8859-1').decode('utf-8'))
dataFinal.Posicion = dataFinal.Posicion.apply(lambda x: bytes(x,'iso-8859-1').decode('utf-8'))
dataFinal.Edad = dataFinal.Edad.apply(lambda x: bytes(x,'iso-8859-1').decode('utf-8'))

dataFinal.Equipo = dataFinal.Equipo.replace({'CA Huracán' : 'Huracán', 
                                             "Newell's Boys" : "Newells Old Boys",
                                             "Gimnasia Plata" : "Gimnasia de la Plata",
                                             "Unión Santa Fe" : "Unión de Santa Fe",
                                             "Martín Tucumán" : "San Martín de Tucumán",
                                             "Atl.Tucumán" : "Atletico Tucumán",
                                             "Defensa Justicia" : "Defensa y Justicia",
                                             "Racing Club" : "Racing"})

for x in range (0, len(dataFinal)):
    if 'cm' in dataFinal.Posicion[x]:
        dataFinal.loc[x, 'Altura'] = dataFinal.Posicion[x]
        
for x in range (0, len(dataFinal)):
    if 'cm' not in dataFinal.Altura[x]:
        dataFinal.loc[x, 'Altura'] = ''

lista = ['Defensa', 'Centrocampista', 'Delantero', 'Portero']

for x in range (0, len(dataFinal)):
    if dataFinal.Nacionalidad[x] in lista:
        dataFinal.loc[x, 'Posicion'] = dataFinal.Nacionalidad[x]

for x in range (0, len(dataFinal)):
    if dataFinal.Nacionalidad[x] in lista:
        dataFinal.loc[x, 'Nacionalidad'] = dataFinal.Provincia[x]
        
dataFinal = dataFinal.drop(['Provincia'], axis=1)

dataFinal['Anio_nacimiento'] = dataFinal.Fecha_Nacimiento.apply(lambda x: x[-4:])
dataFinal['Mes_nacimiento'] = dataFinal.Fecha_Nacimiento.apply(lambda x: x[-7:].split('-')[0])

dataFinal['Posicion_Number'] = dataFinal.Posicion.map({'Portero': 0, 'Defensa': 1, 'Centrocampista': 2, 'Delantero': 3 })

dataFinal['Equipo_Number'] = dataFinal.Equipo.map({'Boca Juniors' : 0,
 'River Plate' : 1,
 'Independiente' : 2,
 'Estudiantes' : 3,
 'San Lorenzo' : 4,
 'Vélez Sarsfield' : 5,
 'Racing' : 6,
 'Huracán' : 7,
 'Newells Old Boys' : 8,
 'Rosario Central' : 9,
 'Gimnasia de la Plata' : 10,
 'Argentinos' : 11,
 'Lanús' : 12,
 'Banfield' : 13,
 'Talleres' : 14,
 'Unión de Santa Fe' : 15,
 'Colón' : 16,
 'Tigre' : 17,
 'Belgrano' : 18,
 'Godoy Cruz' : 19,
 'San Martín de Tucumán' : 20,
 'Atletico Tucumán' : 21,
 'Martín Juan' : 22,
 'Defensa y Justicia' : 23,
 'Aldosivi' : 24,
 'Patronato' : 25})

dataFinal.Edad = dataFinal.Edad.apply(lambda x : x.replace(' años', ''))
dataFinal.Altura = dataFinal.Altura.apply(lambda x : x.replace(' cm', ''))
dataFinal.Altura = dataFinal.Altura.replace('', '0')

dataFinal.Nacionalidad = dataFinal.Nacionalidad.replace({
    'Uruguay, Italia' : 'Uruguay-Italia',
    'Uruguay, Argentina' : 'Uruguay-Argentina',
    'Chile, Argentina' : 'Chile-Argentina',
    'Paraguay, Argentina' : 'Paraguay-Argentina',
    'Santo Tomé y Príncipe' : 'Africa',
})

dataFinal['Nacionalidad_Number'] = dataFinal.Nacionalidad.map({'Argentina' : 0,
 'Paraguay' : 1,
 'Colombia' : 2,
 'Uruguay-Italia' : 3,
 'Chile' : 4,
 'Uruguay' : 5,
 'Uruguay-Argentina' : 6,
 'Chile-Argentina' : 7,
 'Ecuador' : 8,
 'Perú' : 9,
 'Armenia' : 10,
 'Africa' : 11,
 'Paraguay-Argentina' : 12,
 'Venezuela' : 13,
 'Brasil' : 14,
 'Estados Unidos' : 15,
 'España' : 16,
 'Suiza': 17 })

dataFinal['Pais_Number'] = dataFinal.Pais.map({'Argentina' : 0,
 'Paraguay' : 1,
 'Colombia' : 2,
 'Uruguay' : 3,
 'Chile' : 4,
 'Ecuador' : 5,
 'Perú' : 6,
 'Armenia' : 7 ,
 'Santo Tomé y Príncipe' : 8,
 'Venezuela' : 9,
 'Brasil' : 10,
 'Estados Unidos' : 11,
 'España' : 12,
 'Suiza' : 13 })

dataFinal.Altura = dataFinal.Altura.astype('int16')
dataFinal.Edad = dataFinal.Edad.astype('int16')
dataFinal.Anio_nacimiento = dataFinal.Anio_nacimiento .astype('int16')
dataFinal.Mes_nacimiento = dataFinal.Mes_nacimiento.astype('int16')
dataFinal.Posicion_Number = dataFinal.Posicion_Number.astype('int16')
dataFinal.Equipo_Number = dataFinal.Equipo_Number.astype('int16')
dataFinal.Nacionalidad_Number = dataFinal.Nacionalidad_Number.astype('int16')
dataFinal.Pais_Number = dataFinal.Pais_Number.astype('int16')

dataFinal['Pais_abr'] = dataFinal.Pais.map({'Argentina' : 'ARG',
 'Paraguay' : 'PAR',
 'Colombia' : 'COL',
 'Uruguay' : 'URU',
 'Chile' : 'CHI',
 'Ecuador' : 'ECU',
 'Perú' : 'PER',
 'Armenia' : 'ARM' ,
 'Santo Tomé y Príncipe' : 'SANT PRI',
 'Venezuela' : 'VEN',
 'Brasil' : 'BRA',
 'Estados Unidos' : 'EEUU',
 'España' : 'ESP',
 'Suiza' : 'SUI' })

dataFinal['Nac_abr'] = dataFinal.Nacionalidad.map({'Argentina' :'ARG',
 'Paraguay' : 'PAR',
 'Colombia' : 'COL',
 'Uruguay-Italia' : 'URU-ITA',
 'Chile' : 'CHI',
 'Uruguay' : 'URU',
 'Uruguay-Argentina' : 'URU - ARG',
 'Chile-Argentina' : 'CHI - ARG',
 'Ecuador' : 'ECU',
 'Perú' : 'PER',
 'Armenia' : 'ARM',
 'Africa' : 'AFR',
 'Paraguay-Argentina' : 'PAR - URU',
 'Venezuela' : 'VEN',
 'Brasil' : 'BRA',
 'Estados Unidos' : 'EEUU',
 'España' : 'ESP',
 'Suiza' : 'SUI'})

columnas = dataFinal.columns.tolist()

renombre = ['Alt', 'Edad', 'Equip', 'Fech_nac','Nac', 'Nombre', 'Pais', 'Posic', 
            'Año_nac', 'Mes_nac', 'Posic_int', 'Equip_int', 'Nac_int', 'Pais_int', 
            'Pais_abr', 'Nac_abr']

for x in range (0, len(columnas)):
    dataFinal.rename(columns= {columnas[x] : renombre[x]}, inplace=True)
```
