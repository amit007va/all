```python
import requests
import urllib.request
import pandas as pd
import re
import json
import unicodedata
from time import time
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
```

```python
url = 'https://www.tiendanaranja.com/especial-celulares-smartphones.html'
path_file_dest = r'hotSale.json'
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (Platform; Security; OS-or-CPU; Localization; rv:1.4) Gecko/20030624 Netscape/7.1 (ax)'})
```

```python
def write_doc_json(content, mode):
    own_file = open(path_file_dest, mode, encoding='utf-8')
    own_file.write(content.decode('utf-8'))
    own_file.close()
    print('\n•  El archivo json fue sobre-escrito con exito!')
```

```python
if response.status_code == 200:
    info = []
    soup = BeautifulSoup(response.text, "html.parser")
    data = soup.findAll('li', attrs={'class': 'item item-interna'})

    for x in range (0, len(data)):

        nombre_div = data[x].find('h3', {'class': 'product-name'})
        nombre = nombre_div.find('a').get_text()
        url_tel = nombre_div.find('a')['href']
        comercio_vendedor = data[x].find('p', {'class': 'product-vendor'}).get_text()
        precio = data[x].find('span', {'class': 'price'}).get_text().replace(' ', '').replace('\n', ' ').replace('\r', '').replace('\t', '').lstrip().rstrip()
        precio_esp = data[x].find('p', {'class': 'special-price'})
        
        if precio_esp is None:
            precio_especial = precio
        else:
            precio_especial = unicodedata.normalize("NFKD", data[x].find('p', {'class': 'special-price'}).get_text().split('pago')[1].replace(' ', '').replace('\n', ' ').replace('\r', '').lstrip().rstrip())
            cuotas = data[x].find('span', {'class': 'cft-cuotas'}).get_text().replace('\n', ' ').replace('\r', '').replace('\t', '').lstrip().rstrip().split('PTF')[0]
            extras =  unicodedata.normalize("NFKD", data[x].find('span', {'class': 'extras'}).get_text().replace(' ', ' ').lstrip().rstrip())
            precio_final = unicodedata.normalize("NFKD", precio.split(',')[0].replace('.', '').replace('$', '').lstrip())
            precio_descuento = unicodedata.normalize("NFKD", precio_especial.split(',')[0].replace('.', '').replace('$', '').lstrip())
            descuento = int(precio_final) - int(precio_descuento)
            free = data[x].find('img', {'class': 'cucarda_2'})
        
        if free is None:
            envio = 'Pago'
        else:
            envio = 'Gratuito'
            
        response_info = requests.get(url_tel, headers={'User-Agent': 'Mozilla/5.0 (Platform; Security; OS-or-CPU; Localization; rv:1.4) Gecko/20030624 Netscape/7.1 (ax)'})
         
        if response_info.status_code == 200:
            
            soup_dos = BeautifulSoup(response_info.text, "html.parser")
            data_dos = soup_dos.find('table', attrs={'class': 'data-table'})
            
            if data_dos is None:
                
                marca = ''
                vendedor_telefono = ''
                vendedor_email = ''
                vendedor_direccion = ''
            
            else:
                total_adicional = data_dos.findAll('td', attrs={'class': 'data'})
                vendedor_info = soup_dos.find('div', attrs={'class': 'vendor-description'})
                vendor_info_contacto = vendedor_info.findAll('p', {'class' : 'vendor-info first'})
                vendor_info_direccion = vendedor_info.findAll('p', {'class' : 'vendor-info last'})

                vendedor = []
            
                for xi in range (0, len(vendor_info_contacto)):
                    vendedor.append(vendor_info_contacto[xi].get_text().lstrip().rstrip())
                
                for i in range (0, len(vendor_info_direccion)):
                    vendedor.append(vendor_info_direccion[i].get_text().lstrip().rstrip())
        
                info_adicional = []
            
                for j in range (0, len(total_adicional)):
                    info_adicional.append(total_adicional[j].get_text().lstrip())
            
                if len(info_adicional) == 3:
                    marca = info_adicional[2].replace('\n', ' ').replace('\r', '').replace('\t', '').lstrip().rstrip()
                    
                else:
                    marca = info_adicional[0].replace('\n', ' ').replace('\r', '').replace('\t', '').lstrip().rstrip()
            
                if len(vendedor) == 3:
                
                    vendedor_telefono = vendedor[0].replace('\n', ' ').replace('\r', '').replace('\t', '').lstrip().rstrip()
                    vendedor_email = vendedor[1].replace('\n', ' ').replace('\r', '').replace('\t', '').lstrip().rstrip()
                    vendedor_direccion = vendedor[2].replace('\n', ' ').replace('\r', '').replace('\t', '').lstrip().rstrip()
                
                elif len(vendedor) == 2:
                    vendedor_telefono = vendedor[0].replace('\n', ' ').replace('\r', '').replace('\t', '').lstrip().rstrip()
                    vendedor_direccion = vendedor[1].replace('\n', ' ').replace('\r', '').replace('\t', '').lstrip().rstrip()
                
                else:
                    vendedor_telefono = vendedor[0].replace('\n', ' ').replace('\r', '').replace('\t', '').lstrip().rstrip()
                    
        else:
            marca = ''
            vendedor_telefono = ''
            vendedor_email = ''
            vendedor_direccion = ''

        telefono = {
            'nombre' : nombre,
            'precio' : precio,
            'precio_especial' : precio_especial,
            'cuotas' : cuotas,
            'descuento' : descuento,
            'envio' : envio,
            'marca' : marca,
            'comercio_vendedor' : comercio_vendedor,
            'vendedor_telefono' : vendedor_telefono,
            'vendedor_email' : vendedor_email,
            'vendedor_direccion' : vendedor_direccion    
        }
        
        info.append(telefono) 
        
        print('Se guardo el celular n° ' + str(x + 1))
    
content = json.dumps(info, indent=4, sort_keys=True, ensure_ascii=False).encode('utf-8')
write_doc_json(content ,'w+')
data = pd.read_json(path_file_dest, encoding = 'UTF-8')
print('•  Se leyo el json y se creo el dataframe data')
```

```python
data.marca = data.marca.replace('5"','')
data.marca = data.marca.replace('NEGRO','')
data.marca = data.marca.replace('Sí','')
```

```python
data['isDigit'] = data.vendedor_telefono.apply(lambda x: x.isdigit())
```

```python
for x in range(0, len(data)):
    if data.isDigit[x] == False:
        data.loc[x, 'vendedor_telefono'] = ''
```

```python
marca_vacia = data.index[data.marca == ''].tolist()
data.nombre = data.nombre.apply(lambda x : x.lower())

for x in marca_vacia:
    if 'motorola' in data.nombre[x]:
        data.loc[x, 'marca'] = 'motorola'
    elif 'samsung' in data.nombre[x]:
        data.loc[x, 'marca'] = 'motorola'
    elif 'tcl' in data.nombre[x]:
        data.loc[x, 'marca'] = 'tcl'
    elif 'Kanji' in data.nombre[x]:
        data.loc[x, 'marca'] = 'Kanji'
    elif 'blade' in data.nombre[x]:
        data.loc[x, 'marca'] = 'blade'
```
